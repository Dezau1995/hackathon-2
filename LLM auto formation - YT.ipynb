# Code

```python
# Fonction pour rechercher des vidéos sur YouTube en fonction d'un mot-clé.
def search_videos(query, max_results=50):
    videos_search = VideosSearch(query, limit=max_results)
    video_results = videos_search.result()['result']

    video_ids = [video['id'] for video in video_results]
    return video_ids

# Fonction pour obtenir la transcription d'une vidéo.
def get_video_transcript(video_id,languages=['fr']):
    try:
        transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=languages)
        transcript_text = ' '.join([entry['text'] for entry in transcript])
        return transcript_text
    except Exception as e:
        print(f"Erreur lors de la récupération de la transcription pour la vidéo {video_id}: {e}")
        return None
```

```python
# Exemple d'utilisation :
keyword = 'Apprendre python complet'
video_ids = search_videos(keyword)

ids_final = []
transcripts = []
for video_id in video_ids:
    transcript = get_video_transcript(video_id)
    if transcript:
        transcripts.append(transcript)
        ids_final.append(video_id)
        print(f'Transcription pour la vidéo {video_id}:\n{transcript}\n')
```

```python
print(len(transcripts))

print(len(ids_final))
```

```python
urls = ['https://www.youtube.com/watch?v='+ id for id in ids_final]
print(urls)
```

```python
documents = []
for t,url in zip(transcripts,urls):

    doc = Document(
        page_content=t,
        metadata={'url':url}
    )
    documents.append(doc)
```
